logistic regression implementation

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
from sklearn.datasets import make_classification  ## it will create dataset for us and we dont need to do standardization so that we just focus on implementation part
```

```python
## create dataset
x,y=make_classification(n_samples=1000, n_features=10, n_classes=2,random_state=42)
```

```python
pd.DataFrame(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
## model training
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression()
```

```python
logistic.fit(x_train,y_train)
```

```python
y_pred= logistic.predict(x_test)
print(y_pred)
```

```python
logistic.predict_proba(x_test) ## it will give the probability of each class
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
score= accuracy_score(y_test,y_pred)
print("Accuracy of the model is: ", score)
cm= confusion_matrix(y_test,y_pred)
print("Confusion matrix is: ", cm)
print("Classification report is: ", classification_report(y_test,y_pred))
```

hyperparameter tunning and cross validation

```python
model= LogisticRegression()
penalty=['l1', 'l2', 'elsticnet']
c_value=[100,10,1.0,0.1,0.001]
solver=['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
```

```python
parms =dict(penalty= penalty, C= c_value, solver= solver) ## it will create key value pair
```

```python
from sklearn.model_selection import StratifiedKFold
cv= StratifiedKFold() ## it will create 5 folds
```

```python
## grid search Cv
from sklearn.model_selection import GridSearchCV
grid= GridSearchCV(estimator=model, param_grid= parms, scoring='accuracy', cv=cv, n_jobs=-1) ## n_jobs=-1 will use all the processors available in the system
```

```python
grid
```

```python
grid.fit(x_train,y_train) ## it will fit the model on training data
```

```python
grid.best_score_ ## it will give the best score of the model
```

```python
grid.best_params_ ## it will give the best parameters
```

```python
y_pred=grid.predict(x_test) ## it will give the prediction on test data
```

```python
score= accuracy_score(y_test,y_pred)
print("Accuracy of the model is: ", score)
cm= confusion_matrix(y_test,y_pred)
print("Confusion matrix is: ", cm)
print("Classification report is: ", classification_report(y_test,y_pred))
```

Randomised Search Cv

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
model= LogisticRegression()
randomcv= RandomizedSearchCV(estimator=model, param_distributions= parms, scoring='accuracy', cv=5) ## n_iter=10 will take 10 random samples from the parameter space
```

Logistic Regression for multi class classification problem

```python
## create dataset 
x,y= make_classification(n_samples=1000, n_features=10,n_informative=3,n_classes=3,random_state=42)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
##initialise the model
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression(multi_class='ovr')
logistic.fit(x_train,y_train)
y_pred= logistic.predict(x_test)
```

```python
score= accuracy_score(y_test,y_pred)
print("Accuracy of the model is: ", score)
cm= confusion_matrix(y_test,y_pred)
print("Confusion matrix is: ", cm)
print("Classification report is: ", classification_report(y_test,y_pred))
```
